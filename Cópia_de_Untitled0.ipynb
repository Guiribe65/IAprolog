```python
'''/* ****************************************************************************************
* Faculdade de Engenharias Arquitetura e Urbanismo (FEAU) - (Univap)
* Curso: Engenharia da Computação - Data de Entrega: 26/09/2024
* Autor: Guilherme Ribeiro Costa
*
* Turma: 8UNA Disciplina: Aprendizado de Máquina
* Avaliação parcial referente ao 1 - Bimestre
* Observação: <colocar se houver>
*
* guilherme.py
* ****************************************************************************************/'''


import matplotlib.pyplot as plt
import statistics


x = [10, 15, 20, 25, 30, 35, 40, 45, 50]
y = [120, 150, 170, 200, 220, 250, 270, 300, 320]

# Cálculo dos coeficientes da reta de regressão
n = len(x)
m = (n * sum(xi * yi for xi, yi in zip(x, y)) - sum(x) * sum(y)) / (n * sum(xi**2 for xi in x) - sum(x)**2)
b = (sum(y) - m * sum(x)) / n

# Previsão para 37 requisições por segundo
numero_requisicoes = 37
tempo_resposta_previsto = m * numero_requisicoes + b

# Mediana
mediana_x = statistics.median(x)
mediana_y = statistics.median(y)

# Gráfico de dispersão
plt.scatter(x, y, color='blue', label='Dados reais')

# Reta de regressão
requisicoes_segundo = [i for i in range(min(x), max(x) + 1)]
predicao_tempo = [m * xi + b for xi in requisicoes_segundo]
plt.plot(requisicoes_segundo, predicao_tempo, color='red', label='Reta de regressão')

# Destacar a mediana
plt.axvline(mediana_x, color='green', linestyle='--', label='Mediana de x')
plt.axhline(mediana_y, color='purple', linestyle='--', label='Mediana de y')

plt.title('Regressão Linear')
plt.xlabel('Requisições por segundo')
plt.ylabel('Tempo de resposta (ms)')
plt.grid()

plt.show()

# Resultados
print(f"Equação da reta: y = {m:.2f}x + {b:.2f}")
print(f"Previsão do tempo de resposta para 37 requesições por segundo: {tempo_resposta_previsto:.2f} ms")
print(f"Mediana das requisições por segundo: {mediana_x}")
print(f"Mediana do tempo de resposta: {mediana_y}")

```
